```python
df_train_num = df_train.select_dtypes(include = ['int64', 'float64'])
df_train_cat = df_train.select_dtypes(include = ['object'])
df_train_num_cols = df_train_num.drop(columns = 'Premium Amount', axis = 1).columns
df_train_cat_cols = df_train_cat.drop(columns = 'Occupation', axis = 1).columns
extra_cols = ['Occupation']
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

def log_transform(X):
    X = np.log(X + 1)
    return X

df_num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy = 'median')),
    ('scale', StandardScaler())
])


df_cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy = 'most_frequent')),
    ('onehot', OneHotEncoder())
])


df_extra_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy = 'constant', fill_value = 'x')),
    ('onehot', OneHotEncoder())
])


transformer = ColumnTransformer([
    ('num', df_num_pipeline, df_train_num_cols),
    ('cat', df_cat_pipeline, df_train_cat_cols),
    ('extra', df_extra_pipeline, extra_cols)
], remainder = 'passthrough')

X_train = pd.DataFrame(transformer.fit_transform(df_train.drop(columns = 'Premium Amount', axis = 1)), columns = transformer.get_feature_names_out())
X_test = pd.DataFrame(transformer.transform(df_test), columns = transformer.get_feature_names_out())
```

```python
y_train = df_train['Premium Amount']
```

```python
X_train.corrwith(y_train).sort_values(ascending = False)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape = (X_train.shape[1],)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(100, activation = 'leaky_relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(100, activation = 'relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(1, activation = 'linear')
])

optimizer = tf.keras.optimizers.AdamW(learning_rate=0.001, beta_1=0.9, beta_2=0.999, weight_decay=1e-5)
model.compile(loss = 'mean_absolute_error', optimizer = optimizer, metrics = ['mean_absolute_error'])
```

```python
history = model.fit(X_train, y_train, epochs = 25, validation_split = 0.2)
```
